```python
import json
import requests
def chunk_text(text,max_len=3000):
    for i in range(0,len(text),max_len):
        yield text[i:i+max_len]
```

```python
def sending_chat_message_to_google_chats(title,description=None,path=None):
    chunk = list(chunk_text(description)) if description else []
    widgets=[]
    if chunk:
        for id,part in enumerate(chunk):
            widgets.append({
                "keyValue" : {
                    "topLabel" : "Description",
                    "content" : part,
                    "contentMultiline": True
                }
            })
    else:
        widgets.append({
            "keyValue" : "Description",
            "content" : "No description provided"
        })
    widgets.append({
        "keyValue" : {
        "topLabel" : "Notebook Path",
        "content" : path or "No path provided"
        }
    })
    if title and (description is None and path is None):
        payload = {
        "text" : f"{title}"
    }
    else:
        payload = {
            "cards": [
                {
                    "header": {
                        "title": "Databricks Job Notification",
                        "subtitle": f'❌ {title}'
                    },
                    "sections": [
                        {
                            "widgets": widgets
                        }
                    ]
                }
            ]
        }
    headers = {'Content-Type': 'application/json; charset=UTF-8'}
    try:
        # webhook_url = dbutils.secrets.get(scope="google_chat", key="dl_notification")
        webhook_url = 'https://chat.googleapis.com/v1/spaces/AAQAyga_3hc/messages?key=AIzaSyDdI0hCZtE6vySjMm-WEfRq3CPzqKqqsHI&token=9uWfp5uDEwuhJVRBZ-iuRuMWTFeKbB8KAyZw4jYWb5o'
        response = requests.post(webhook_url,data=json.dumps(payload),headers=headers)
        return response.status_code
    except Exception as e:
        return str(e)
```
